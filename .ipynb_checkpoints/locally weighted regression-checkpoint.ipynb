```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("tips_10.csv")
bill = data.total_bill
tip = data.tip
```

```python
lowess = sm.nonparametric.lowess(tip, bill, frac=0.3)
x_smooth, y_smooth = lowess.T
```

```python
plt.scatter(bill,tip, color = "green")
plt.plot(x_smooth, y_smooth, color = "red", linewidth=5)
plt.xlabel = "Total Bill"
plt.ylabel = "Tip"
plt.show()
```
